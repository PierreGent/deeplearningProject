```python
# import library
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import cv2

# Width, Height ...
img_width=100
img_height=100
```

```python
# load train dataframe
df_train = pd.read_csv('./dog-breed-identification/labels.csv')

# print 10 first rows
df_train.head(10)
```

```python
# Here, we show the number of image for each breed in the dataset

plot_class_distribution = df_train['breed'].value_counts().plot(kind='barh', fontsize='15', figsize=(20, 50))
plot_class_distribution.set(xlabel="Images per class", ylabel="Classes")
plot_class_distribution.xaxis.label.set_size(20)
plot_class_distribution.yaxis.label.set_size(20)
plt.show()
```
